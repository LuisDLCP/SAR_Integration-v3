```python
import numpy as np
import pandas as pd
```

```python
f=lambda x:np.exp(x)*np.cos(x)
```

```python
# Hallando la integral por el metodo de Gaus  n=3
I=5/9*f(-0.7746)+8/9*f(0)+5/9*f(0.7746)
#
print(f"El valor de la integral es: {I}")
```

```python
I3=0.5*(np.exp(1)*(np.sin(1)+np.cos(1))-np.exp(-1)*(np.cos(1)-np.sin(1)))
#
print(f"El valor de la integral es: {I3}")
```

```python
# Calculando los errores
e1=np.abs((I-I3)/I3*100)
print(f"El error por cuadratura de gauss n=3 es: {e1}")
#
e2=np.abs((I2-I3)/I3*100)
print(f"El error por cuadratura de gauss n=4 es: {e2}")
```

## 2.a)

```python
# Paso h=0.2
h=0.2
x=np.arange(0,4+h,h)
f=lambda x,y: x-2*x*y
```

```python
#Hallando los vectores Y & F
y=[0]
F=[]
for i in range(len(x)):
    F.append(f(x[i],y[i]))
    y_next = y[i]+h*F[i]
    y.append(y_next)
# Convirtiendo a df
df = pd.DataFrame(list(zip(x,y,F)), columns=['x','y','f(x,y)'])
df
```

```python
# Paso h=0.05
h=0.05
x=np.arange(0,4+h,h)
f=lambda x,y: x-2*x*y
```

```python
#Hallando los vectores Y & F
y=[0]
F=[]
for i in range(len(x)):
    F.append(f(x[i],y[i]))
    y_next = y[i]+h*F[i]
    y.append(y_next)
# Convirtiendo a df
df = pd.DataFrame(list(zip(x,y,F)), columns=['x','y','f(x,y)'])
df
```

## 2.c)

```python
#Hallando la tabla
h=0.2
x=np.arange(0,4+h,h)
f=lambda x,y: x-2*x*y

y=[0]
k1=[]
k2=[]
k3=[]
k4=[]
for i in range(len(x)):
    k1.append(f(x[i],y[i]))
    k2.append(f(x[i]+h/2,y[i]+h/2*k1[i]))
    k3.append(f(x[i]+h/2,y[i]+h/2*k2[i]))
    k4.append(f(x[i]+h,y[i]+h*k3[i]))
    y_next = y[i]+h/6*(k1[i]+2*k2[i]+2*k3[i]+k4[i])
    y.append(y_next)
# Convirtiendo a df
df = pd.DataFrame(list(zip(x,y,k1,k2,k3,k4)), columns=['x','y','k1','k2','k3','k4'])
df
```
